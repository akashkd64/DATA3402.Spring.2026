*Exercise 3:* Modify Exercise 2 so that it takes a matrix in the format from Exercise 1 and draws a tic‑tac‑toe board with `"X"`s and `"O"`s.

````python
# Write your solution here
def tic_tac_board(game_board):
    rows = len(game_board)
    cols = len(game_board[0])

    for i in range(rows):
        # print top line of each row
        print(" ---" * cols)

        line = ""
        for j in range(cols):
            cell = game_board[i][j]

            if cell == 0:
                sign = " "
            elif cell == 1:
                sign = "X"
            else:
                sign = "O"

            line = line + "| " + sign + " "

        line = line + "|"
        print(line)
    print(" ---" * cols)
````

*Exercise 4:* Write a function that takes an `n×n` matrix representing a tic‑tac‑toe game and returns one of the following values:

- `-1` if the game is **incomplete** (still empty spaces and no winner)
- `0` if the game is a **draw**
- `1` if **player 1** (`"X"`) has won
- `2` if **player 2** (`"O"`) has won

Here are some example inputs you can use to test your code:


````python
# Write your solution here
def test_winner(tictac_board):
    n = len(tictac_board)
    
    for r in range(n):
        first = tictac_board[r][0]
        if first != 0:
            all_match = True
            for c in range(1, n):
                if tictac_board[r][c] != first:
                    all_match = False
                    break
            if all_match:
                return first
    for c in range(n):
        first = tictac_board[0][c]
        if first != 0:
            all_match = True
            for r in range(1, n):
                if tictac_board[r][c] != first:
                    all_match = False
                    break
            if all_match:
                return first
    first = tictac_board[0][0]
    if first != 0:
        all_match = True
        for i in range(1, n):
            if tictac_board[i][i] != first:
                all_match = False
                break
        if all_match:
            return first
    first = tictac_board[0][n - 1]
    if first != 0:
        all_match = True
        for i in range(1, n):
            if tictac_board[i][n - 1 - i] != first:
                all_match = False
                break
        if all_match:
            return first

    for r in range(n):
        for c in range(n):
            if tictac_board[r][c] == 0:
                return -1  

    return 0           
````

*Exercise 11:* Develop a version of the game where one player is the computer.

Note: you do **not** need an extensive search for the best move. For example, you can have the computer:
- block obvious losses
- otherwise try to create a winning row/column/diagonal


````python
def computer_turn(board, computer_player):
    n = len(board)

    for i in range(n):
        for j in range(n):
            if board[i][j] == 0:
                board[i][j] = computer_player
                if test_winner(board) == computer_player:
                    return
                board[i][j] = 0

    human_player = 3 - computer_player
    for i in range(n):
        for j in range(n):
            if board[i][j] == 0:
                board[i][j] = human_player
                if test_winner(board) == human_player:
                    board[i][j] = computer_player
                    return
                board[i][j] = 0


    mid = n // 2
    if board[mid][mid] == 0:
        board[mid][mid] = computer_player
        return


    corners = [(0,0), (0,n-1), (n-1,0), (n-1,n-1)]
    for r, c in corners:
        if board[r][c] == 0:
            board[r][c] = computer_player
            return

    for i in range(n):
        for j in range(n):
            if board[i][j] == 0:
                board[i][j] = computer_player
                return


def play_game_vs_computer(board):
    current_player = 1  

    while True:
        new_board(board)

        if current_player == 1:
            player_turn(board, 1)
        else:
            print("Computer thinking...")
            computer_turn(board, 2)

        winner = test_winner(board)

        if winner == 1 or winner == 2:
            new_board(board)
            print(f"Player {winner} wins!")
            break

        if winner == 0:
            new_board(board)
            print("It's a draw!")
            break

        current_player = 2 if current_player == 1 else 1
````

*Exercise 12:* Develop a version of the game where one player is the computer. This time, write a computer player using exhaustive search with a max depth parameter, similar to lecture.

````python
# Write your solution here
def evaluate(board, maximizing_player):
    winner = test_winner(board)

    if winner == maximizing_player:
        return 10
    elif winner == 3 - maximizing_player:
        return -10
    elif winner == 0:
        return 0

    my_marks = sum(cell == maximizing_player for row in board for cell in row)
    opp_marks = sum(cell == (3 - maximizing_player) for row in board for cell in row)
    return my_marks - opp_marks


def minimax(board, depth, max_depth, current_player, maximizing_player):
    winner = test_winner(board)

    if winner != -1 or depth == max_depth:
        return evaluate(board, maximizing_player)

    n = len(board)

    if current_player == maximizing_player:
        best_score = -999
        for r in range(n):
            for c in range(n):
                if board[r][c] == 0:
                    board[r][c] = current_player
                    score = minimax(board, depth + 1, max_depth,
                                    3 - current_player, maximizing_player)
                    board[r][c] = 0
                    best_score = max(best_score, score)
        return best_score
    else:
        best_score = 999
        for r in range(n):
            for c in range(n):
                if board[r][c] == 0:
                    board[r][c] = current_player
                    score = minimax(board, depth + 1, max_depth,
                                    3 - current_player, maximizing_player)
                    board[r][c] = 0
                    best_score = min(best_score, score)
        return best_score


def computer_turn(board, computer_player=2, max_depth=6):
    n = len(board)
    best_score = -999
    best_move = None

    for r in range(n):
        for c in range(n):
            if board[r][c] == 0:
                board[r][c] = computer_player
                score = minimax(board, 0, max_depth,
                                3 - computer_player, computer_player)
                board[r][c] = 0

                if score > best_score:
                    best_score = score
                    best_move = (r, c)

    if best_move is not None:
        r, c = best_move
        board[r][c] = computer_player


def play_vs_computer(n=3, max_depth=6):
    board = [
    [0,0,0],
    [0,0,0],
    [0,0,0]]
    current_player = 1

    while True:
        new_board(board)
        winner = test_winner(board)

        if winner != -1:
            if winner == 1:
                print("You win!")
            elif winner == 2:
                print("Computer wins!")
            else:
                print("It's a draw!")
            break

        if current_player == 1:
            player_turn(board, 1)
        else:
            print("Computer thinking...")
            computer_turn(board, 2, max_depth)

        current_player = 3 - current_player
````

*Exercise 13:* Make the 2 computer players play each-other for 10 games on a 3x3, then 4x4, then 5x5 grid. Set the max depth so that the games only take seconds. Measure the "smarter" player's win rate for each grid.

````python
# Write your solution here
def make_board(n):
    return [[0]*n for _ in range(n)]

def status(b):
    n=len(b)
    for r in b:
        if r[0] and all(x==r[0] for x in r): return r[0]
    for c in range(n):
        v=b[0][c]
        if v and all(b[r][c]==v for r in range(n)): return v
    v=b[0][0]
    if v and all(b[i][i]==v for i in range(n)): return v
    v=b[0][n-1]
    if v and all(b[i][n-1-i]==v for i in range(n)): return v
    if all(x for r in b for x in r): return 0
    return -1

def moves(b):
    n=len(b)
    return [(i,j) for i in range(n) for j in range(n) if b[i][j]==0]

def evalb(b, smart):
    w=status(b)
    if w==smart: return 10
    if w==3-smart: return -10
    if w==0: return 0
    my=sum(x==smart for r in b for x in r)
    op=sum(x==3-smart for r in b for x in r)
    return my-op

def minimax(b, p, smart, d, D):
    w=status(b)
    if w!=-1 or d==D: return evalb(b, smart)
    best=-10**9 if p==smart else 10**9
    for i,j in moves(b):
        b[i][j]=p
        s=minimax(b, 3-p, smart, d+1, D)
        b[i][j]=0
        best=max(best,s) if p==smart else min(best,s)
    return best

def best_move(b, p, smart, D):
    best=None
    bests=-10**9 if p==smart else 10**9
    for i,j in moves(b):
        b[i][j]=p
        s=minimax(b, 3-p, smart, 0, D)
        b[i][j]=0
        if (p==smart and s>bests) or (p!=smart and s<bests):
            bests=s; best=(i,j)
    return best

def play_game(n, smart=1, Dsmart=5, Ddumb=2):
    b=make_board(n); p=1
    while True:
        D=Dsmart if p==smart else Ddumb
        i,j=best_move(b,p,smart,D)
        if (i,j) is None: return 0
        b[i][j]=p
        w=status(b)
        if w!=-1: return w
        p=3-p

def run_tournament(sizes=(3,4,5), games=10):
    for n in sizes:
        Dsmart,Ddumb = (9,5) if n==3 else ((4,2) if n==4 else (3,1))
        s=d=dr=0
        for _ in range(games):
            w=play_game(n, smart=1, Dsmart=Dsmart, Ddumb=Ddumb)
            if w==1: s+=1
            elif w==2: d+=1
            else: dr+=1
        print(f"{n}x{n}: Smart {s}/{games} ({100*s/games:.0f}%), Dumb {d}, Draw {dr} | depths {Dsmart}/{Ddumb}")
````

````python
# Test your solution here
run_tournament((3,4,5), games=2)
````
